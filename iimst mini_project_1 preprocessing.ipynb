```python
import pandas as pd

df = pd.read_csv("5e5eb7bd-b059-45ea-b165-7b2faf386a61_Data.csv")
df.head()
```

```python
df = df.drop(columns=["Country Code", "Series Code"])
df.head()
```

```python
years = [str(year) for year in range(2000, 2024)]

df = df[["Country Name", "Series Name"] + 
        [col for col in df.columns if any(year in col for year in years)]]

df.head()
```

```python
df_long = df.melt(
    id_vars=["Country Name", "Series Name"],
    var_name="Year",
    value_name="Value"
)

df_long.head()
```

```python
df_long["Year"] = df_long["Year"].str.extract('(\d+)')
df_long["Year"] = df_long["Year"].astype(int)
```

```python
# Replace '..' with NaN
df_long["Value"] = df_long["Value"].replace("..", pd.NA)

# Convert to numeric
df_long["Value"] = pd.to_numeric(df_long["Value"], errors="coerce")
```

```python
df_final = df_long.pivot_table(
    index=["Country Name", "Year"],
    columns="Series Name",
    values="Value"
).reset_index()

df_final.head()
```

```python
df_final.isna().sum()
```
